```python
import pandas as  pd
import re
import collections
```

```python
from pyecharts.charts import WordCloud
from pyecharts import options as opts
```

```python
data = pd.read_excel('C:/Users/wo/Desktop/去除好评的分词抽取/jialidechouquwuzheng.xlsx')
```

```python
data.head()
```

```python
string_data=[]
```

```python
for i in data['位置']:
    if i!='[]':
        string_data.append(i)
for i in data['服务']:
    if i!='[]':
        string_data.append(i)
for i in data['价格']:
    if i!='[]':
        string_data.append(i)
for i in data['环境']:
    if i!='[]':
        string_data.append(i)
for i in data['食物']:
    if i!='[]':
        string_data.append(i)
```

```python
string_data
```

```python
# 5.进行词频统计，使用pyecharts生成词云
# 词频统计
word_counts = collections.Counter(string_data)  # 对分词做词频统计
word_counts_top = word_counts.most_common(80)  # 获取前100最高频的词
```

```python
# 绘图
# https://gallery.pyecharts.org/#/WordCloud/wordcloud_custom_mask_image
# 去pyecharts官网找模板代码复制出来修改
c = (
    WordCloud()
    .add("", word_counts_top)#根据词频最高的词
    .render("jialidewordcloud.html")#生成页面
)
```
