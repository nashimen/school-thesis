```python
"""
POS标注，使用百度LAC库
"""
import pandas as pd
from LAC import LAC
```

```python
data = pd.read_excel('../data/data_cleaned.xlsx')
data
```

```python
# POS报错：TypeError: object of type 'float' has no len()
# 数据中存在NaN数据
# data[~ data['评论文本'].apply(lambda x: isinstance(x, str))]
data = data[~ data['评论文本'].isna()]
data.shape
```

```python
# baidu/LAC词性标注
lac = LAC(mode='lac')
data['LAC'] = data['评论文本'].apply(lambda x: lac.run(x))
data
```

```python
data['word'], data['part'] = data['LAC'].apply(lambda x: x[0]), data['LAC'].apply(lambda x: x[1])
data
```

```python
# 去停用词
with open('../data/hit_stopwords.txt', 'r', encoding='utf-8') as f:
    stop_words = f.readlines()
stop_words = [stop_word.strip('\n') for stop_word in stop_words]
data['stoped_word'] = data['word'].apply(lambda x: list(filter(lambda word: word not in stop_words, x)))
data
```

```python
def get_nouns(lac):
    word, part = lac[0], lac[1]
    nouns = []
    for word_token, part_token in zip(word, part):
        if part_token == 'n':
            nouns.append(word_token)
    return nouns


data['nouns'] = data['LAC'].apply(get_nouns)
data
```

```python
data['word2vec_corpus'] = data['stoped_word'].apply(lambda x: ' '.join(x))
data
```

```python
data['nouns_corpus'] = data['nouns'].apply(lambda x: ' '.join(x))
data
```

```python
data.to_excel('../data/data_POS_corpora.xlsx', index=False)
```

```python
data['word2vec_corpus'].to_csv('../data/word2vec_corpus.txt', index=False)
```

```python
data['nouns_corpus'].to_csv('../data/nouns.txt', index=False, header=False)
```

```python
data['word'].apply(lambda x: ' '.join(x)).to_csv('../data/word2vec_corpus_with_stoped_punctuation.txt', index=False, header=False)
```
