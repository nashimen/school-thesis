```python
'''
sentence-transformer embedding
model:paraphrase-multilingual-mpnet-base-v2
'''
from sentence_transformers import SentenceTransformer
import pandas as pd

# 本地导入模型指定模型路径
model = SentenceTransformer('E:\OneDrive - tgblx\Python\model\paraphrase-multilingual-mpnet-base-v2')
```

```python
data = pd.read_excel('../data/mywv_wwm_tencent_count100.xlsx')
data
```

```python
data['sentence_transformer'] = data['noun'].apply(lambda x: model.encode(x))
data
```

```python
data['sentence_transformer'] = data['sentence_transformer'].apply(lambda x: ' '.join([str(number) for number in x]))
data
```
